```python
import pandas as pd
# Load train set:
df = pd.read_parquet('/home/ndo/vardict_ML/processed_data.parquet')
# Shuffle the train dataset
df = df.sample(frac=1).reset_index(drop=True)
df['VD'] = df['VD'].fillna(df['VD'].mean())
# Define X (features) and y (target)
X_train = df.drop(columns=['labels'])  # Features
y_train = df['labels']  # Target labels

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform the training data
X_train_normalized = scaler.fit_transform(X_train)
X_train_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)

# Step 1: Combine X_train and y_train into a single DataFrame
train_normalized_data = X_train_normalized.copy()  # Make a copy of X_train to avoid modifying the original
train_normalized_data['labels'] = y_train  # Add y_train as a column named 'labels' in the DataFrame

# Step 2: Write the DataFrame to a Parquet file
# train_normalized_data.to_parquet('/home/ndo/vardict_ML/process_train_normalized_dataset.parquet')

# print("Train dataset saved to 'train_scaled_dataset.parquet'")

train_normalized_data
```
