```python
import pandas as pd
# Load train set:
df = pd.read_parquet('/home/ndo/vardict_ML/processed_data_test_106.parquet')
# Shuffle the train dataset
df = df.sample(frac=1).reset_index(drop=True)
df['VD'] = df['VD'].fillna(df['VD'].mean())
# Define X (features) and y (target)
X_test = df.drop(columns=['labels'])  # Features
y_test = df['labels']  # Target labels

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform the training data
X_test_normalized = scaler.fit_transform(X_test)
X_test_normalized = pd.DataFrame(X_test_normalized, columns=X_test.columns)

# Step 1: Combine X_train and y_train into a single DataFrame
test_normalized_data = X_test_normalized.copy()  # Make a copy of X_train to avoid modifying the original
test_normalized_data['labels'] = y_test  # Add y_train as a column named 'labels' in the DataFrame

# Step 2: Write the DataFrame to a Parquet file
# train_normalized_data.to_parquet('/home/ndo/vardict_ML/process_train_normalized_dataset.parquet')

# print("Train dataset saved to 'train_scaled_dataset.parquet'")

test_normalized_data
```
